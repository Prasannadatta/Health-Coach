```python
import random
import json
import os
import time
import requests

# =========================
# USDA CONFIG + CACHING
# =========================

BASE_URL = "https://api.nal.usda.gov/fdc/v1/foods/search"
DETAIL_URL = "https://api.nal.usda.gov/fdc/v1/food"

API_KEY = "YTu6rNh3VVltd0pGa3XgzvcXiMxbWU0ohawDTWD5"  # <-- set this!

# Cache: food_name -> (calories_per_gram, protein_per_gram)
FOOD_CACHE = {}


def get_food_calories(food_name, weight_grams=100):
    """
    Return (calories_per_gram, protein_per_gram) for a given food name.

    - Uses in-memory cache so repeated ingredients are instant
    - Uses fewer retries and smaller sleeps to avoid super slow runs
    - If anything fails, falls back to (0.0, 0.0) so the pipeline never crashes
    """
    # ---------- Cache hit ----------
    if food_name in FOOD_CACHE:
        return FOOD_CACHE[food_name]

    # ---------- 1) Search for the food ----------
    params = {
        "query": food_name,
        "api_key": API_KEY,
        "pageSize": 1,
    }

    data = None
    for attempt in range(3):  # fewer retries
        try:
            r = requests.get(BASE_URL, params=params, timeout=10)
        except Exception as e:
            print(f"âŒ Search request error for {food_name}: {e}")
            time.sleep(0.2 * (attempt + 1))
            continue

        if r.status_code == 200 and "application/json" in r.headers.get("Content-Type", ""):
            try:
                data = r.json()
            except Exception as e:
                print(f"âŒ JSON decode error for {food_name}: {e}")
                data = None
            break

        # non-200 or non-JSON â†’ backoff and retry
        time.sleep(0.2 * (attempt + 1))

    if not data or "foods" not in data or len(data["foods"]) == 0:
        print("âŒ Search failed or no foods for:", food_name)
        FOOD_CACHE[food_name] = (0.0, 0.0)
        return 0.0, 0.0

    fdc_id = data["foods"][0]["fdcId"]

    # ---------- 2) Get detailed nutrients ----------
    detail_params = {"api_key": API_KEY}
    detail_data = None

    for attempt in range(3):
        try:
            r2 = requests.get(f"{DETAIL_URL}/{fdc_id}", params=detail_params, timeout=10)
        except Exception as e:
            print(f"âŒ Detail request error for {food_name}: {e}")
            time.sleep(0.2 * (attempt + 1))
            continue

        if r2.status_code == 200 and "application/json" in r2.headers.get("Content-Type", ""):
            try:
                detail_data = r2.json()
            except Exception as e:
                print(f"âŒ JSON decode error (detail) for {food_name}: {e}")
                detail_data = None
            break

        time.sleep(0.2 * (attempt + 1))

    if not detail_data:
        print("âŒ Detail fetch failed for:", food_name)
        FOOD_CACHE[food_name] = (0.0, 0.0)
        return 0.0, 0.0

    # ---------- Parse nutrients ----------
    calories = 0.0
    proteins = 0.0

    for nutrient in detail_data.get("foodNutrients", []):
        if not isinstance(nutrient, dict):
            continue
        nutrient_info = nutrient.get("nutrient", {})
        nid = nutrient_info.get("id")

        # 1008 = Energy (kcal)
        if nid == 1008:
            val = nutrient.get("amount", nutrient.get("value", 0.0)) or 0.0
            calories = float(val)

        # 1003 = Protein (g)
        if nid == 1003:
            val = nutrient.get("amount", nutrient.get("value", 0.0)) or 0.0
            proteins = float(val)

    # Most USDA nutrients are per 100g
    cal_per_g = calories / 100.0 if calories else 0.0
    pro_per_g = proteins / 100.0 if proteins else 0.0

    FOOD_CACHE[food_name] = (cal_per_g, pro_per_g)
    return cal_per_g, pro_per_g


# =========================
# LABEL NORMALIZATION
# =========================

def normalize_label(val, default=""):
    """
    Turn weird label formats into a clean single string:
    - ["american"]       -> "american"
    - "american, italian"-> random choice of one
    - "lunch/dinner"     -> random choice of one
    """
    if isinstance(val, list):
        flat = [str(x).strip() for x in val if str(x).strip()]
        return random.choice(flat) if flat else default

    if isinstance(val, str):
        s = val.strip()

        # Try JSON array like '["american"]'
        if (s.startswith("[") and s.endswith("]")) or (s.startswith("{") and s.endswith("}")):
            try:
                parsed = json.loads(s)
                return normalize_label(parsed, default)
            except Exception:
                pass

        # Split on common separators
        for sep in [",", "/", ";", "|"]:
            if sep in s:
                parts = [p.strip() for p in s.split(sep) if p.strip()]
                return random.choice(parts) if parts else default

        return s or default

    return default


# =========================
# QA PAIR GENERATOR
# =========================

def create_qa_pairs_with_ingredient_calories(
    df,
    num_samples=100,
    output_path=None,   # e.g. "qa_pairs.jsonl"
    save_every=5,       # flush to disk every N examples
    append=False        # if True, append to existing file
):
    """
    Streaming QA generator.

    Each QA pair:
        {
          "question": "...",
          "answer": {
            "recipe_name": str,
            "calories": float,   # per serving (from ingredients)
            "proteins": float,   # per serving (from ingredients)
            "ingredients": [
              {
                "ingredient": str,
                "weight_g_or_ml": float,
                "calories_per_unit": float,   # kcal per g
                "total_calories": float,      # kcal in that ingredient
                "protein": float              # g protein in that ingredient
              }, ...
            ]
          }
        }

    - Questions are diverse (calorie, diet/health, protein, mixed).
    - Writes to JSONL so progress is never lost.
    - Also returns a Python list of all pairs.
    """
    qa_pairs = []

    f = None
    if output_path is not None:
        mode = "a" if append and os.path.exists(output_path) else "w"
        f = open(output_path, mode, encoding="utf-8")

    try:
        for i in range(num_samples):
            row = df.sample(1).iloc[0]

            recipe_name = row.get("recipe_name", "A healthy recipe")
            meal_type_raw = row.get("meal_type", "meal")
            cuisine_raw   = row.get("cuisine_type", "")
            diet_raw      = row.get("diet_labels", "")
            health_raw    = row.get("health_labels", "")

            meal_type = normalize_label(meal_type_raw, "meal")
            cuisine   = normalize_label(cuisine_raw, "any")
            diet      = normalize_label(diet_raw, "balanced")
            health    = normalize_label(health_raw, "healthy")

            # ---- Servings ----
            servings_val = row.get("servings", 1)
            try:
                servings = max(1, int(servings_val))
            except Exception:
                servings = 1

            # ---- Ingredients list ----
            ingredients_raw = row.get("ingredients", [])
            if isinstance(ingredients_raw, str):
                try:
                    ingredients_list = json.loads(ingredients_raw)
                except Exception:
                    ingredients_list = []
            else:
                ingredients_list = ingredients_raw or []

            ingredients_with_cal = []
            total_cals_all = 0.0
            total_pro_all = 0.0

            for item in ingredients_list:
                # Normalize ingredient entry
                if isinstance(item, dict):
                    food_name = (
                        item.get("food")
                        or item.get("ingredient")
                        or "ingredient"
                    )
                    weight = float(
                        item.get("weight")
                        or item.get("weight_g_or_ml")
                        or 0.0
                    )
                else:
                    food_name = str(item)
                    weight = 0.0

                if weight <= 0:
                    continue

                # Assumes get_food_calories returns (cal_per_gram, protein_per_gram)
                cal_per_g, pro_per_g = get_food_calories(food_name, weight_grams=weight)

                calories_per_unit = cal_per_g
                total_calories = calories_per_unit * weight
                protein_total = pro_per_g * weight

                total_cals_all += total_calories
                total_pro_all += protein_total

                ingredients_with_cal.append({
                    "ingredient": food_name,
                    "weight_g_or_ml": round(weight, 2),
                    "calories_per_unit": round(calories_per_unit, 2),
                    "total_calories": round(total_calories, 2),
                    "protein": round(protein_total, 2),
                })

            if servings <= 0:
                servings = 1

            calories_per_serv = round(total_cals_all / servings, 2)
            protein_per_serv = round(total_pro_all / servings, 2)

            # --------- QUESTION GENERATION (DIVERSE) ---------
            question_family = random.choice(
                ["calorie", "diet_health", "protein", "mixed"]
            )

            # make sure limit > actual per-serving calories
            cal_limit = max(
                int(calories_per_serv) + random.randint(20, 200),
                int(calories_per_serv) + 20,
            )

            if question_family == "calorie":
                templates = [
                    "Suggest a {cuisine} {meal} under {cal} calories per serving.",
                    "What is a good {meal} in {cuisine} cuisine under {cal} calories per serving?",
                    "Give me a low-calorie {cuisine} {meal} under {cal} calories per serving.",
                ]
                tmpl = random.choice(templates)
                question = tmpl.format(
                    cuisine=cuisine,
                    meal=meal_type,
                    cal=cal_limit,
                )

            elif question_family == "diet_health":
                templates = [
                    "Find a {cuisine} recipe that is {diet} and {health} and under {cal} calories per serving.",
                    "Recommend a {diet}, {health} {cuisine} dish under {cal} calories per serving.",
                    "Show me a {health} {cuisine} {meal} that stays under {cal} calories per serving.",
                ]
                tmpl = random.choice(templates)
                question = tmpl.format(
                    cuisine=cuisine,
                    meal=meal_type,
                    diet=diet,
                    health=health,
                    cal=cal_limit,
                )

            elif question_family == "protein":
                # keep protein target realistic relative to this recipe
                protein_target = max(3, int(protein_per_serv * random.uniform(0.7, 1.2)))
                templates = [
                    "Recommend a {meal} with at least {protein} g protein per serving.",
                    "I need a {meal} that has at least {protein} g of protein per serving.",
                    "Give me a high-protein {meal} (â‰¥ {protein} g protein per serving).",
                ]
                tmpl = random.choice(templates)
                question = tmpl.format(
                    meal=meal_type,
                    protein=protein_target,
                )

            else:  # "mixed"
                templates = [
                    "Give me a {diet} {meal} in {cuisine} style under {cal} calories per serving.",
                    "Suggest a {cuisine} {meal} that fits a {diet} diet and stays below {cal} calories per serving.",
                    "What {diet} {meal} from {cuisine} cuisine can I eat under {cal} calories per serving?",
                ]
                tmpl = random.choice(templates)
                question = tmpl.format(
                    cuisine=cuisine,
                    meal=meal_type,
                    diet=diet,
                    cal=cal_limit,
                )

            # --------- ANSWER (schema stays identical) ---------
            answer = {
                "recipe_name": recipe_name,
                "calories": calories_per_serv,
                "proteins": protein_per_serv,
                "ingredients": ingredients_with_cal,
            }

            qa = {"question": question, "answer": answer}
            qa_pairs.append(qa)

            # --------- Streaming save to JSONL ---------
            if f is not None:
                f.write(json.dumps(qa, ensure_ascii=False) + "\n")
                if (i + 1) % save_every == 0:
                    f.flush()
                    os.fsync(f.fileno())
                    print(f"ðŸ’¾ Saved {i+1} examples to {output_path}")

        # final flush
        if f is not None:
            f.flush()
            os.fsync(f.fileno())
            print(f"âœ… Finished. Total saved examples: {num_samples}")

    finally:
        if f is not None:
            f.close()

    return qa_pairs
```

```python
from datasets import load_dataset  # unused if you stick to CSV, but kept for compatibility
import random
import json
import pandas as pd
import time
import requests

# -------------------------------------------------
# Load the dataset
# -------------------------------------------------
# If you want HF:
# ds = load_dataset("datahiveai/recipes-with-nutrition")
# df = ds["train"].to_pandas()

# Your current CSV path (adjust if needed)
df = pd.read_csv("./../data/recipes-with-nutrition.csv")

# -------------------------------------------------
# USDA API setup
# -------------------------------------------------
API_KEY = "YTu6rNh3VVltd0pGa3XgzvcXiMxbWU0ohawDTWD5"  # <-- PUT YOUR REAL KEY HERE
BASE_URL = "https://api.nal.usda.gov/fdc/v1/foods/search"


# ---------- USDA LOOKUP (UNCHANGED LOGIC) ----------

def get_food_calories(food_name, weight_grams):
    """
    Return (cal_per_gram, protein_per_gram) for a food using USDA API.
    LOGIC KEPT THE SAME AS YOUR VERSION.
    """
    params = {
        "query": food_name,
        "api_key": API_KEY,
        "pageSize": 1
    }
    data = None
    for attempt in range(8):
        r = requests.get(BASE_URL, params=params)
        if r.status_code == 200 and "application/json" in r.headers.get("Content-Type", ""):
            data = r.json()
            print("âœ… Search OK:", food_name)
            break
        time.sleep(0.3 * (attempt + 1))

    if not data or "foods" not in data or len(data["foods"]) == 0:
        print("âŒ Search failed:", food_name)
        return 0.0, 0.0

    fdc_id = data["foods"][0]["fdcId"]
    print(fdc_id)

    # Food details
    detail_url = f"https://api.nal.usda.gov/fdc/v1/food/{fdc_id}?api_key={API_KEY}"
    detail_data = None
    for attempt in range(8):
        resp = requests.get(detail_url)
        if resp.status_code == 200 and "application/json" in resp.headers.get("Content-Type", ""):
            detail_data = resp.json()
            print("âœ… DATA Fetch:", food_name)
            break
        time.sleep(0.3 * (attempt + 1))

    if not detail_data:
        return 0.0, 0.0

    calories = 0.0
    proteins = 0.0

    # USDA can be two shapes: (nutrientId/value) or (nutrient: {id}, amount)
    for nutrient in detail_data.get("foodNutrients", []):
        if not isinstance(nutrient, dict):
            continue

        nid = nutrient.get("nutrientId")
        if nid is None:
            nut_obj = nutrient.get("nutrient", {}) or {}
            nid = nut_obj.get("id")

        value = nutrient.get("value")
        if value is None:
            value = nutrient.get("amount", 0.0)

        if nid == 1008:      # Energy
            calories = float(value)
        elif nid == 1003:    # Protein
            proteins = float(value)

    print("raw calories (per serving):", calories)
    print("raw proteins (per serving):", proteins)

    serving_size = detail_data.get("servingSize")
    if not serving_size or serving_size == 0:
        # fall back: assume values are per 100 g
        cal_per_gram = calories / 100.0
        pro_per_gram = proteins / 100.0
    else:
        cal_per_gram = calories / float(serving_size)
        pro_per_gram = proteins / float(serving_size)

    print("final per gram c:", cal_per_gram)
    print("final per gram p:", pro_per_gram)
    print("=======================================")
    return cal_per_gram, pro_per_gram


# ---------- QUESTION TEMPLATES (DIVERSE) ----------

calorie_templates_under = [
    "Give me a {diet} {meal} under {cal} calories.",
    "Suggest a {cuisine} recipe below {cal} calories.",
    "What can I eat for {meal} under {cal} calories?",
    "Recommend a {diet} friendly {meal} with less than {cal} calories.",
    "I need a {meal} option with calories under {cal}.",
    "Find me something {cuisine} style under {cal} calories.",
    "What is a good {meal} choice below {cal} calories?",
    "Show me a low-calorie {meal} under {cal} calories.",
    "Show me a {health} recipe under {cal} calories.",
    "Suggest a {health} friendly {meal} under {cal} calories.",
    "Give me a {health} and {diet} option below {cal} calories.",
    "Find a {cuisine} recipe that is {health} and under {cal} calories.",
    "Recommend a {meal} that is {health} and under {cal} calories.",
    "I'm trying to stay under {cal} calories â€” what {meal} would you suggest?",
    "Help me choose a {diet} {meal} under {cal} calories in {cuisine} cuisine.",
    "Suggest a light {meal} under {cal} calories that still feels satisfying.",
    "What is a simple {cuisine} {meal} I can have under {cal} calories?",
]

calorie_templates_above = [
    "Give me a {diet} {meal} above {cal} calories.",
    "Suggest a hearty {cuisine} recipe over {cal} calories.",
    "What can I eat for {meal} with more than {cal} calories?",
    "Recommend a higher-calorie {meal} above {cal} calories.",
    "Find me a {cuisine} style dish with calories above {cal}.",
    "Show me a filling {meal} thatâ€™s over {cal} calories.",
    "Suggest a {health} recipe with more than {cal} calories.",
    "I need a more filling {meal} with at least {cal} calories.",
    "Whatâ€™s a hearty {diet} {meal} above {cal} calories?",
    "Recommend a satisfying {cuisine} dish over {cal} calories.",
]

calorie_templates_exact = [
    "Give me a {diet} {meal} around {cal} calories.",
    "Suggest a {cuisine} recipe thatâ€™s approximately {cal} calories.",
    "Recommend a {meal} close to {cal} calories.",
    "Find a {health} friendly {meal} thatâ€™s about {cal} calories.",
    "Show me a {cuisine} recipe near {cal} calories.",
    "Iâ€™m aiming for about {cal} calories â€” what {meal} fits that?",
    "What is a {diet} {meal} roughly {cal} calories in {cuisine} cuisine?",
]

all_calorie_templates = (
    [("under", t) for t in calorie_templates_under] +
    [("above", t) for t in calorie_templates_above] +
    [("exact", t) for t in calorie_templates_exact]
)


# ---------- LABEL NORMALIZATION HELPERS ----------

def clean_str(s: str) -> str:
    """Strip whitespace + stray quotes/brackets from a label."""
    return s.strip().strip("[]'\" ").strip()


def normalize_label(value, default=""):
    """
    Turn meal_type / cuisine_type / diet_labels into a single clean string.

    Handles:
      - plain strings: "american"
      - CSV strings: "Vegetarian, Dairy-Free"
      - JSON-list strings: '["american", "mexican"]'
      - Python lists: ["american", "mexican"]
    """
    # List-like â†’ choose one
    if isinstance(value, list):
        candidates = [clean_str(str(v)) for v in value if clean_str(str(v))]
        return random.choice(candidates) if candidates else default

    if isinstance(value, str):
        s = value.strip()
        if not s:
            return default

        # JSON-like list as string
        if s.startswith("[") and s.endswith("]"):
            try:
                parsed = json.loads(s)
                return normalize_label(parsed, default)
            except Exception:
                # fall through to CSV-like handling
                pass

        # CSV-like string
        if "," in s:
            parts = [clean_str(p) for p in s.split(",") if clean_str(p)]
            return random.choice(parts) if parts else default

        # simple single label, just clean
        return clean_str(s) or default

    return default


def normalize_label_list(value):
    """
    For health_labels: always return a clean list of strings.
    Accepts list, CSV string, or JSON-list string.
    """
    if isinstance(value, list):
        return [clean_str(str(v)) for v in value if clean_str(str(v))]

    if isinstance(value, str):
        s = value.strip()
        if not s:
            return []
        # JSON-like list string
        if s.startswith("[") and s.endswith("]"):
            try:
                parsed = json.loads(s)
                return normalize_label_list(parsed)
            except Exception:
                pass
        # CSV-like
        return [clean_str(p) for p in s.split(",") if clean_str(p)]

    return []


def r2(x):
    """Round safely to 2 decimals (handle None)."""
    if x is None:
        return None
    return round(float(x), 2)


# ---------- MAIN QA GENERATOR (SAVE LOGIC KEPT) ----------

def create_qa_pairs_with_ingredient_calories(df, num_samples=100, food_api_func=None):
    """
    df: DataFrame with recipe and ingredient info

    - Ingredient calories/proteins are per serving (based on USDA per-gram values).
    - Top-level calories & proteins are sums of ingredients per serving.
    - All nutrition numbers are rounded to 2 decimals.
    - Saving logic is the same: qa_pairs.json is overwritten in each loop.
    """
    qa_pairs = []

    for _ in range(num_samples):

        row = df.sample(1).iloc[0]

        recipe = row.get("recipe_name", "a healthy dish")

        # --- servings handling (robust) ---
        servings_value = row.get("servings", 1)
        try:
            servings = int(servings_value)
        except (TypeError, ValueError):
            servings = 1
        if servings <= 0:
            servings = 1  # avoid divide-by-zero

        # ------- labels (normalized) -------
        meal_raw    = row.get("meal_type", "meal")
        diet_raw    = row.get("diet_labels", "regular")
        cuisine_raw = row.get("cuisine_type", "any")
        health_raw  = row.get("health_labels", "")

        meal    = normalize_label(meal_raw, "meal")
        diet    = normalize_label(diet_raw, "regular")
        cuisine = normalize_label(cuisine_raw, "any")
        health_list = normalize_label_list(health_raw)

        # convert ingredients field
        ingredients_raw = row.get("ingredients", [])
        if isinstance(ingredients_raw, str):
            try:
                ingredients_list = json.loads(ingredients_raw)
            except Exception:
                ingredients_list = []
        else:
            ingredients_list = ingredients_raw or []

        ingredients_with_cal = []
        total_cal_recipe = 0.0
        total_pro_recipe = 0.0

        for item in ingredients_list:
            if isinstance(item, dict):
                food_name = item.get("food", "") or item.get("ingredient", "") or "ingredient"
                weight = item.get("weight", 0.0)
            else:
                food_name = str(item)
                weight = 0.0

            if weight is None:
                weight = 0.0

            # ðŸ”’ keep your API call logic
            cal_per_unit, pro_per_unit = get_food_calories(food_name, weight)

            ingredient_total_cal = cal_per_unit * weight
            ingredient_total_pro = pro_per_unit * weight

            total_cal_recipe += ingredient_total_cal
            total_pro_recipe += ingredient_total_pro

            cal_per_serv = ingredient_total_cal / servings
            pro_per_serv = ingredient_total_pro / servings

            ingredients_with_cal.append({
                "ingredient": food_name,
                "weight_g_or_ml": r2(weight),
                "calories_per_unit": r2(cal_per_unit),
                "total_calories": r2(cal_per_serv),   # per serving
                "protein": r2(pro_per_serv),          # per serving
            })

        calories_per_serving = total_cal_recipe / servings if servings > 0 else 0.0
        protein_per_serving  = total_pro_recipe / servings if servings > 0 else 0.0

        # -------- pick some health labels for the question text --------
        selected_health = []
        if health_list:
            p = random.random()
            if p < 0.60:
                selected_health = [random.choice(health_list)]
            elif p < 0.90:
                k = min(2, len(health_list))
                selected_health = random.sample(health_list, k)
            else:
                k = min(3, len(health_list))
                selected_health = random.sample(health_list, k)
        health_str = ", ".join(selected_health) if selected_health else "healthy"

        # -------- choose template + {cal} --------
        tmpl_type, q_template = random.choice(all_calorie_templates)

        base_cal = int(calories_per_serving) if calories_per_serving > 0 else random.randint(200, 700)

        if tmpl_type == "under":
            cal_for_q = random.randint(base_cal + 20, base_cal + 200)
        elif tmpl_type == "above":
            lower = max(50, base_cal - 200)
            upper = max(lower + 20, base_cal - 20)
            cal_for_q = random.randint(lower, upper)
        else:  # "exact"
            delta = random.randint(10, 50)
            lower = max(50, base_cal - delta)
            upper = max(lower + 10, base_cal + delta)
            cal_for_q = random.randint(lower, upper)

        question = q_template.format(
            diet=diet,
            meal=meal,
            cuisine=cuisine,
            cal=cal_for_q,
            health=health_str,
        )

        answer = {
            "recipe_name": recipe,
            "calories": r2(calories_per_serving),
            "proteins": r2(protein_per_serving),
            "ingredients": ingredients_with_cal,
        }

        qa_pairs.append({
            "question": question,
            "answer": answer
        })

        # ðŸ”’ SAVE LOGIC LEFT AS YOU WROTE IT
        save_path = "qa_pairs.json"
        if save_path is not None:
            with open(save_path, "w") as f:
                json.dump(qa_pairs, f, indent=4)

    return qa_pairs


# -------------------------------------------------
# Run generator
# -------------------------------------------------
if __name__ == "__main__":
    # change num_samples to whatever you want (just remember USDA is slow)
    qa_output = create_qa_pairs_with_ingredient_calories(df, num_samples=1000)
    print(f"Generated {len(qa_output)} QA pairs")
```
