```python
import random
import json
import time

calorie_templates = [
    "Give me a {diet} {meal} under {cal} calories.",
    "Suggest a {cuisine} recipe below {cal} calories.",
    "What can I eat for {meal} under {cal} calories?",
    "Recommend a {diet} friendly {meal} with less than {cal} calories.",
    "I need a {meal} option with calories under {cal}.",
    "Find me something {cuisine} style under {cal} calories.",
    "What is a good {meal} choice below {cal} calories?",
    "Show me a low-calorie {meal} under {cal} calories.",
    "Show me a {health} recipe under {cal} calories.",
    "Suggest a {health} friendly {meal} under {cal} calories.",
    "Give me a {health} and {diet} option below {cal} calories.",
    "Find a {cuisine} recipe that is {health} and under {cal} calories.",
    "Recommend a {meal} that is {health} and under {cal} calories.",
]

# Use your existing function
BASE_URL = "https://api.nal.usda.gov/fdc/v1/foods/search"

def get_food_calories(food_name, weight_grams):
    # Search food
    search_url = f"https://api.nal.usda.gov/fdc/v1/foods/search?api_key={API_KEY}&query={food_name}&pageSize=1"

    params = {
        "query": food_name,
        "api_key": API_KEY,
        "pageSize": 1
    }
    data = None
    for attempt in range(8):  # retry up to 5 times
        r = requests.get(BASE_URL, params=params)

        # ✅ valid JSON response
        if r.status_code == 200 and "application/json" in r.headers.get("Content-Type",""):
            data= r.json()
            print("✅ Search OK:", food_name)
            # print(data)
            break   

        # ✅ USDA rate limit or timeout
        time.sleep(0.3 * (attempt + 1))
    

    
    if not data or "foods" not in data or len(data["foods"]) == 0:
        print("❌ Search failed:", food_name)
        return None
    
    fdc_id = data["foods"][0]["fdcId"]
    print(fdc_id)

    # Get food details
    detail_url = f"https://api.nal.usda.gov/fdc/v1/food/{fdc_id}?api_key={API_KEY}"
    for attempt in range(8):  # retry up to 5 times
        detail_data = requests.get(detail_url)

        # ✅ valid JSON response
        if detail_data.status_code == 200 and "application/json" in detail_data.headers.get("Content-Type",""):
            detail_data= detail_data.json()
            print("✅ DATA Fetch:", food_name)
            # print(data)
            break   

        # ✅ USDA rate limit or timeout
        time.sleep(0.3 * (attempt + 1))
    
    # print(detail_data)

    calories = 0
    # print(detail_data["foodNutrients"])
    for nutrient in detail_data.get("foodNutrients", []):
        # print(nutrient)
        if isinstance(nutrient, dict):
            nutrient_info = nutrient.get("nutrient", {})
            if nutrient_info.get("id") == 1008:
                calories = nutrient.get("amount", 0)
                if calories is None:
                    calories = nutrient.get("value")
        
                print(calories)
                break
            
    proteins = 0        
    for nutrient in detail_data.get("foodNutrients", []):
        # print(nutrient)
        if isinstance(nutrient, dict):
            nutrient_info = nutrient.get("nutrient", {})
            if nutrient_info.get("id")== 1003:
                proteins = nutrient.get("amount")
                if proteins is None:
                    proteins = nutrient.get("value")
                break

    # print(detail_data)
    
    print("calories",calories)
    print("proteins",proteins)
    # serving_size=detail_data.get("servingSize",{})
    # if serving_size is None or serving_size=={}:
    #     serving_size = 100
    # elif serving_size == 0:
    #     serving_size = 1
            
    # print("serving_size",serving_size)
    print("final per unit c:",calories/100)
    print("final per unit p:",proteins/100)
    print("=======================================")
    return calories/100,proteins/100



def create_qa_pairs_with_ingredient_calories(df, num_samples=100, food_api_func=None):
    """
    df: DataFrame with recipe and ingredient info
    food_api_func: function that takes ingredient name and returns USDA food JSON
    """
    qa_pairs = []

    for _ in range(num_samples):

        row = df.sample(1).iloc[0]

        recipe = row.get("recipe_name", "a healthy dish")
        recipe_cal = row.get("calorie_count", random.randint(200, 700))

        # Generate a calorie limit above recipe calories
        try:
            recipe_cal_val = int(recipe_cal)
        except:
            recipe_cal_val = random.randint(200, 700)

        cal_limit = random.randint(recipe_cal_val + 20, recipe_cal_val + 200)

        meal = row.get("meal_type", "meal")
        diet = row.get("diet_labels", "regular")
        cuisine = row.get("cuisine_type", "any")
        serving_size_gms = row.get("serving_size_gms", None)
        servings_value = row.get("servings", 0)   # maybe string, maybe number
        servings = int(servings_value)
        ingredients_raw = row.get("ingredients", [])
        Protein_gms_per_serving=row.get("Protein_gms_per_serving")
        
        print("ingredients_raw")
        print(ingredients_raw)

        # Convert ingredient string → list
        if isinstance(ingredients_raw, str):
            ingredients_list = json.loads(ingredients_raw)   # now this is a Python list
        else:
            ingredients_list = ingredients_raw 

        print(ingredients_list)
        cal=0
        final_cal=0
        ingredients_with_cal=[]

        for item in ingredients_list:
            cal,pro = get_food_calories(item["food"], item["weight"])
            print("food",item["food"])
            print("weight",item["weight"])
            
            
            weight=item["weight"]
            cal_per_unit=cal
            
            final_pro=(pro*weight)
            print("final_pro")
            print(final_pro)
            final_cal=(cal*weight)
            print("final_cal")
            print((cal*weight))
            

            ingredients_with_cal.append({
                "ingredient": item if isinstance(item, str) else item.get("food"),
                "weight_g_or_ml": weight,
                "calories_per_unit": cal_per_unit,
                "total_calories": final_cal,
                "protein": final_pro
            })

            print(ingredients_with_cal)

        health_labels = row.get("health_labels", [])
        if isinstance(health_labels, str):
            health_list = [h.strip() for h in health_labels.split(",") if h.strip()]
        else:
            health_list = []

        # Pick random health labels
        selected_health = []
        p = random.random()
        if len(health_list) > 0:
            if p < 0.70:
                selected_health = [random.choice(health_list)]
            elif p < 0.95:
                selected_health = random.sample(health_list, min(2, len(health_list)))
            else:
                k = min(4, len(health_list))
                selected_health = random.sample(health_list, random.randint(3, k))
        health_str = ", ".join(selected_health) if selected_health else ""

        q_template = random.choice(calorie_templates)
        question = q_template.format(
            diet=diet,
            meal=meal,
            cuisine=cuisine,
            cal=cal_limit,
            health=health_str
        )

        answer = {
            "recipe_name": recipe,
            "calories": recipe_cal/servings,
            "proteins": Protein_gms_per_serving,
            "ingredients": ingredients_with_cal
        }

        qa_pairs.append({
            "question": question,
            "answer": answer
        })

    return qa_pairs


# Example usage:
# qa_output = create_qa_pairs_with_ingredient_calories(df, num_samples=3000, food_api_func=calories_per_gram_or_ml)
qa_output = create_qa_pairs_with_ingredient_calories(df, num_samples=2)
with open("qa_pairs.json", "w") as f:
     json.dump(qa_output, f, indent=4)
```
